```python
from qiskit import QuantumCircuit
import numpy as np

def custom_qft(circuit, n):
    for i in range(n):
        # Apply Hadamard gate to the current qubit
        circuit.h(i)
        # Apply controlled-phase rotations to the remaining qubits
        for j in range(i + 1, n):
            # The angle is 2*pi / 2^(j-i+1) 
            angle = np.pi / (2**(j - i))
            circuit.cp(angle, j, i) 
        # Optional: Add a barrier between steps 
        #circuit.barrier()
    
    # Apply final SWAP gates to reverse the order of qubits 
    for i in range(n // 2): 
        circuit.swap(i, n - 1 - i) 
    return circuit 

# Create a new circuit 
qc_manual = QuantumCircuit(3) 
custom_qft(qc_manual, 3) 

# Draw the circuit 
display(qc_manual.draw('mpl',fold=-1))
```
