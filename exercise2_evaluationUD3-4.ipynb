```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Math, display
```

```python
# PARAMETERS AND INITIAL CONDITIONS

# number of time steps (iterations) --> j
jmax=10     # the exercise just ask for 3, but to have it more complete we do more
j=np.linspace(0, jmax, jmax+1)  # time steps array

# available positions in the chain
p=np.linspace(-jmax, jmax, 2*jmax+1)  # positions only from -j to j

#probabilty amplitudes
P=np.zeros((len(p), len(j)))  # probability amplitude array initialized to zero

# wave function arrays initialization
phi_up=np.zeros((len(p), len(j)), dtype=complex)  # wave function array initialized to zero
phi_down=np.zeros((len(p), len(j)), dtype=complex)  # wave function array initialized to zero
phi_up[jmax]=1/np.sqrt(2)  # initial condition: all amplitude at position 0, half spin up
phi_down[jmax]=1j/np.sqrt(2) # initial condition: all amplitude at position 0, half spin down
```

```python
# EVOLUTION RULES

print("Showing only the results different from 0:")
for j in range(0, jmax-1):  # loop over time steps
   print(f"Time step j={j}:")
   for p in range(0, 2*jmax):  # loop over positions
      phi_up[p,j+1]=1/np.sqrt(2)*(phi_up[p-1,j]+phi_down[p+1,j])  # evolution rule for spin up
      phi_down[p,j+1]=1/np.sqrt(2)*(phi_up[p-1,j]-phi_down[p+1,j])  # evolution rule for spin down
      P[p, j]=np.abs(phi_up[p, j])**2+np.abs(phi_down[p, j])**2  # calculate probability amplitude
       # SHOWING RESULTS
      if phi_up[p,j] != 0:
         print(f"Position p={p-jmax}:")
         latex_str = fr"\psi_{{up}}={phi_up[p,j]:.3g}"
         display(Math(latex_str))
      if phi_down[p,j] != 0:
         latex_str = fr"\psi_{{down}}={phi_down[p,j]:.3g}"
         display(Math(latex_str))
      if P[p,j] != 0:
         latex_str = fr"P={P[p,j]:.3g}"
         display(Math(latex_str))
   # plotting the probability distribution at each time step
   positions = np.arange(len(P[:, j])) - jmax
   plt.figure(figsize=(6, 3))
   plt.bar(positions, P[:, j], width=0.8, color='skyblue', alpha=0.7)  # nicer color
   plt.title(f'Probability Distribution, Time Step j={j}')
   plt.xlabel('Position p')
   plt.ylabel('Probability P')
   plt.ylim(0, 1)
   plt.grid(True, linestyle='--', alpha=0.5)
   plt.xticks(positions)  # ensures correct x-ticks
   plt.show()
   print("\n")


```
